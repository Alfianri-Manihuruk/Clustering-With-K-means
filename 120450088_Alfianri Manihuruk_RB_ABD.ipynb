```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

### B. Memanggil dan Memahami Data

Langkah selanjutnya, yaitu meng-input dataset. Dataset yang akan digunakan pada tulisan ini 
adalah dataset Advertising. data ini merupakan sekumpulan data Advertising
Berikut adalah langkah untuk meng-input dataset dan melihat detail informasi tentang dataset

```python
df  = pd.read_csv("Advertising.csv")
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
# ploting data awal
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(df['TV'], df['Radio'], label = 'radio')
plt.scatter(df['TV'], df['Newspaper'], label = 'newspaper')
plt.scatter(df['TV'], df['Sales'], label = 'Sales' )
plt.title("Ploting Nilai Advertising", fontsize=20, color='blue')

plt.xlabel('TV', fontsize= 15, color= 'black')
plt.ylabel('Radio, Sales, Newspaper', fontsize= 15, color= 'black')
plt.legend()
plt.show()
```

### C. Prediksi

pada tahap ini kita akan melakukan prediksi untuk untuk melakukan clustering dari data yang kita miliki.
Nah, hal terpenting adalah tidak hanya menghitung y_predicted saja melainkan memberi label kembali datanya. 
Percuma saja jika kita tidak mampu memetakan kembali siapa saja yang masuk kategori klaster 0 dan 1.

```python
km = KMeans(n_clusters=2)
y_predicted =km.fit_predict(df[['TV','Radio', 'Newspaper', 'Sales']])
y_predicted
```

```python
# membuat dan menampilkan data hasil clustering tadi
df['klaster'] = y_predicted
df.head(10)
```

## D. Finishing

pada tahap ini kita kembali memvisualisasikan data hasil clustering tadi dalam bentuk grafik 

### 1. Visualisasi atribut TV dan Radio

```python
df1 = df[df.klaster == 0]
df2 = df[df.klaster == 1]

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(df1['TV'], df1['Radio'], label= 'klaster 1', c= 'green')
plt.scatter(df2['TV'], df2['Radio'], label= 'klaster 2', c= 'blue')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c='red', s= 100, marker= 'o', label='centroid')
plt.legend(loc = 1)
plt.xlabel('TV', fontsize= 15, color= 'black')
plt.ylabel('Radio', fontsize= 15, color= 'black')

plt.title("Hasil Klaster Kmeans", fontsize= 20, color= 'blue')
```

### 2. Visualisasi atribut TV dan Sales

```python
df1= df[df.klaster ==0]
df2= df[df.klaster ==1]
figure(num= None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(df1['TV'], df1['Sales'], label= 'klaster 1', c= 'red')
plt.scatter(df2['TV'], df2['Sales'], label= 'klaster 2', c= 'blue')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], c='black', s= 150, marker= 'o', label='centroid')
plt.legend(loc=2)

plt.xlabel('TV', fontsize= 15, color= 'black')
plt.ylabel('Sales', fontsize= 15, color= 'black')
plt.title("Hasil Klaster Kmeans", fontsize= 20, color= 'blue')
```
